```python
import pandas

data = pandas.read_csv('../datasets/dt_data.csv.csv')
images = list(data['Изображения'])

my_dict = []
"""
{
    'last': url,
    'new_original': url,
    'new_432': url,
    'new_128': url,
}
"""
domain = 'https://deshevle-tut.ru/data/photos/'
for i in range(len(images)):
    if type(images[i]) != float:
        _images = images[i].split(', ')
        img_name = _images[0].split('/')[-1]
        my_dict.append( {
            'last': _images[0],
            'new_original': domain + 'original/' + img_name,
            'new_432': domain + '432/432__' + img_name,
            'new_128': domain + '128/128__' + img_name,
        })
    else:
        my_dict.append({
            'last': None,
            'new_original': None,
            'new_432': None,
            'new_128': None,
        })
my_dict
```

```python
my_data = pandas.DataFrame(my_dict)

my_data.to_csv('my_csv.csv')
```
